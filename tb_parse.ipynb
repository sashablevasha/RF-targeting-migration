```python
from tbparse import SummaryReader
import numpy as np
import matplotlib.pyplot as plt
```

```python
log_dir = "./runs/2025.05.27.15.42.55/"
reader = SummaryReader(log_dir)
df = reader.scalars
log_dir = "./runs/2025.05.27.16.40.08"
reader = SummaryReader(log_dir)
df2 = reader.scalars
```

```python
plt.plot(np.arange(0,121),df[df["tag"]=='entropy/entropy']["value"],label="   ",ls=":",c="black")
plt.plot(np.arange(121,121+len(df2[df2["tag"]=='entropy/entropy'])),df2[df2["tag"]=='entropy/entropy']["value"],label="   PPO",ls="-",c="black")
plt.legend()
plt.title("    ")
plt.xlabel("")
plt.ylabel(" ")
plt.grid()
plt.savefig("./output/entropy.png")
```

```python
plt.plot(np.arange(0,121),df[df["tag"]=='rewards/rewards']["value"],label="   ",ls=":",c="black")
plt.plot(np.arange(121,121+len(df2[(df2["tag"]=='rewards/rewards')])),df2[df2["tag"]=='rewards/rewards']["value"],label="   PPO",ls="-",c="black")
plt.legend()
plt.title("        ")
plt.xlabel("")
plt.ylabel("      ")
plt.grid()
plt.savefig("./output/rewards.png")
```
