```python
from tbparse import SummaryReader
import numpy as np
import matplotlib.pyplot as plt
```

```python
log_dir = "./runs/2025.05.27.15.42.55/"
reader = SummaryReader(log_dir)
df = reader.scalars
log_dir = "./runs/2025.05.27.16.40.08"
reader = SummaryReader(log_dir)
df2 = reader.scalars
```

```python
plt.plot(np.arange(0,121),df[df["tag"]=='entropy/entropy']["value"],label="Предобучение на экспертных траекториях",ls=":",c="black")
plt.plot(np.arange(121,121+len(df2[df2["tag"]=='entropy/entropy'])),df2[df2["tag"]=='entropy/entropy']["value"],label="Дообучение с помощью PPO",ls="-",c="black")
plt.legend()
plt.title("Энтропия стратегии по мере обучения")
plt.xlabel("Эпоха")
plt.ylabel("Значение энтропии")
plt.grid()
plt.savefig("./output/entropy.png")
```

```python
plt.plot(np.arange(0,121),df[df["tag"]=='rewards/rewards']["value"],label="Предобучение на экспертных траекториях",ls=":",c="black")
plt.plot(np.arange(121,121+len(df2[(df2["tag"]=='rewards/rewards')])),df2[df2["tag"]=='rewards/rewards']["value"],label="Дообучение с помощью PPO",ls="-",c="black")
plt.legend()
plt.title("Среднее из суммы наград по батчу по мере обучения")
plt.xlabel("Эпоха")
plt.ylabel("Значение среднего из суммы наград по батчу")
plt.grid()
plt.savefig("./output/rewards.png")
```
